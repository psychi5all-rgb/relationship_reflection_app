```python
import spacy
import streamlit
import sklearn
import nltk
import pandas
import numpy
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "I felt hurt when they ignored what I said."

doc = nlp(text)

for token in doc:
    print(token.text, token.pos_)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import nltk
nltk.download("vader_lexicon")

analyzer = SentimentIntensityAnalyzer()

text = "I felt frustrated because they ignored me."

print(analyzer.polarity_scores(text))
```

```python
from empath import Empath

lexicon = Empath()

text = "I was angry because they ignored what I said."

print(lexicon.analyze(text, normalize=True))
```
