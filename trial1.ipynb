```python
with open("data/cornell_dialog/movie_lines.txt", "r", encoding="latin-1") as f:
    lines = f.readlines()

print(lines[:5])
```

```python
texts = []

for line in lines:
    
    parts = line.split(" +++$+++ ")
    
    if len(parts) == 5:
        texts.append(parts[4].strip())

print(texts[:10])
```

```python
import re

clean_texts = []

for t in texts:
    
    t = re.sub(r"\s+", " ", t)
    
    clean_texts.append(t.strip())

print(clean_texts[:10])
```

```python
len(clean_texts)
```

```python
texts = []

for line in lines:
    
    sentences = line.split("__eou__")
    
    for s in sentences:
        
        s = s.strip()
        
        if len(s) > 2:
            texts.append(s)

print(texts[:10])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

vectorizer = TfidfVectorizer(max_features=5000)

X = vectorizer.fit_transform(texts)

labels = ["neutral"] * len(texts)   # temporary labels

model = LogisticRegression(max_iter=200)

model.fit(X, labels)
```

```python
keywords = [
    "always",
    "never",
    "you don't",
    "you never",
    "why did you",
    "stop",
    "listen",
    "not true",
    "that's wrong",
    "sorry",
    "I didn't mean",
]

filtered = []

for t in candidate_texts:
    
    lower = t.lower()
    
    if any(k in lower for k in keywords):
        filtered.append(t)

print(len(filtered))
print(filtered[:20])
```
